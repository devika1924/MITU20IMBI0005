```python
import matplotlib.pyplot as plt
from abc import ABC, abstractmethod
from Bio import SeqIO
from Bio.SeqUtils import GC

class BioData(ABC):
    def __init__(self, name, data):
        self.name = name
        self.data = data

    @abstractmethod
    def analyze(self):
        pass

    @abstractmethod
    def visualize(self):
        pass

class ProteinSequence(BioData):
    def __init__(self, name, sequence):
        super().__init__(name, sequence)

    def analyze(self):
        print("\nAnalyzing Protein Sequence...")
        print("Name: " + self.name)
        print("Length of the sequence: " + str(len(self.data)))
        print("Unique Amino acids: " + str(len(set(self.data))))

    def visualize(self):
        print("\nVisualizing Protein Sequence...")
        amino_acid_count = {aa: self.data.count(aa) for aa in set(self.data)}
        plt.bar(amino_acid_count.keys(), amino_acid_count.values())
        plt.title("Amino Acid Distribution")
        plt.show()

class DnaSequence(BioData):
    def __init__(self, name, sequence):
        super().__init__(name, sequence)

    def analyze(self):
        print("\nAnalyzing DNA Sequence...")
        print("Name: " + self.name)
        print("Length of the sequence: " + str(len(self.data)))
        print("Unique Nucleotides: " + str(len(set(self.data))))
        
        # Calculate GC content
        gc_content = GC(self.data)
        print(f"GC Content: {gc_content:.2f}%")

    def visualize(self):
        print("\nVisualizing DNA Sequence...")
        nucleotide_count = {nt: self.data.count(nt) for nt in set(self.data)}
        plt.bar(nucleotide_count.keys(), nucleotide_count.values())
        plt.title("Nucleotide Distribution")
        plt.show()
        
def menu():
    while True:
        print("\nBioinformatics Data Analysis")
        print("1. Analyze Protein Sequence")
        print("2. Visualize Protein Sequence")
        print("3. Analyze DNA Sequence")
        print("4. Visualize DNA Sequence")
        print("5. Exit")
        
        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            name = input("Enter the name of the protein: ")
            sequence = input("Enter the protein sequence: ")
            protein = ProteinSequence(name, sequence)
            protein.analyze()
        elif choice == '2':
            name = input("Enter the name of the protein: ")
            sequence = input("Enter the protein sequence: ")
            protein = ProteinSequence(name, sequence)
            protein.visualize()
        elif choice == '3':
            name = input("Enter the name of the DNA: ")
            sequence = input("Enter the DNA sequence: ")
            dna = DnaSequence(name, sequence)
            dna.analyze()
        elif choice == '4':
            name = input("Enter the name of the DNA: ")
            sequence = input("Enter the DNA sequence: ")
            dna = DnaSequence(name, sequence)
            dna.visualize()
        elif choice == '5':
            print("Exited the program.")
            break
        else:
            print("Invalid choice, please try again.")

menu()
```
